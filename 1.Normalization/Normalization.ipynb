```python
#importing library 
import pandas as pd 
import numpy as np
```

```python
def logCPM(a):
    #indexing in a variable for the dataset
    id1=a.index
    #selecting those columns which have only numeric value
    b=a.select_dtypes(include=['int64'])
    #taking only numeric value for cpm calculation
    a=a._get_numeric_data()
    #converting dataframe into numpy array
    a=a.values
    try:
        #cpm(count per millions ) normalization where converting rows into million with respect to mean 
        cpm=[]
        for i in range(0,len(a)):
                colsum=a[:,i].sum()

                c1=(a[:,i]/colsum)*10**6
                cpm.append(c1)
    
    except:
        print("ignore error")
    #converting cpm lists into array
    cpm=np.array(cpm)
    #converting cpm value into log value of base 2
    lgv=np.log2(cpm+1)  #cpm+1 is a call of psuedo count which helps avoiding to get infinite values
    #converting logcpm array to dataframe
    logcpm=pd.DataFrame(lgv).T
    #now replacing logcpm column with those numerical column of the dataset
    logcpm.columns=b.columns
    #indexing genename of logcpm
    logcpm['gene']=id1
    lk=logcpm.set_index('gene')
    return lk
    
```

```python
data=pd.read_csv('GSE145011_A549_WT_KO_readcounts.txt.gz',sep="\t")
data
```

```python
#calling function logCPM
logCPM(data)
```

```python
lg=logCPM(data)
```

```python
#calculating zscore value of logcpm
import scipy.stats as st
zscr=st.zscore(lg)
    
zscr=pd.DataFrame(zscr)
new=zscr.T
#fill null value with zero
zscr=new.fillna(0)
zscr=zscr.values
zscr
```

```python
#variance value calculation of each column
var = []
for i in range(0,len(zscr)):
    vari = zscr[:,i].var()
    var.append(vari)
```
